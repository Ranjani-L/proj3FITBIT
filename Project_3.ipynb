```python
continious=['Age','Weight (kg)', 'Height (m)', 'Max_BPM',
       'Avg_BPM', 'Resting_BPM', 'Session_Duration (hours)',
       'Fat_Percentage', 'Water_Intake (liters)',
       'BMI', 'Base_MET',
       'HR_Intensity', 'Effective_MET', 'Calories_Burned (kcal)']
```

```python
cat=['Gender','Workout_Frequency (days/week)', 'Workout_Type','Experience_Level']
```

```python
from scipy import stats
```

```python
#2 sample T test
#cont vs cont
def continious_vs_continious(col1,col2):
    t=0
    f=0
    for i in range(31):
        sample1 = df[col1].sample(frac=0.03)
        sample2 = df[col2].sample(frac=0.03)
        t_test , p_value = stats.ttest_ind(sample1,sample2)
        if p_value<0.05:
            f=f+1
        else:
            t=t+1
    if t>f:
        return True
    else:
        return False
```

```python
continious_vs_continious('Age','BMI')
```

```python
#continious(col1) category(col2)
def continious_vs_category(col1,col2):
    group = df[col2].unique()
    data={}
    for i in group:
        data[i] = df[col1][df[col2]==i]
    f_value, p_value = stats.f_oneway(*[i for i in data.values()])
    return False if p_value < 0.05 else True
```

```python
continious_vs_category('BMI','Gender')
```

```python
#cat_vs_cat
def cat_vs_cat(col1,col2):
    return True if stats.chi2_contingency(pd.crosstab(df[col1],df[col2]))[1] < 0.05 else False
```

```python
cat_vs_cat('Gender','Workout_Type')
```

```python
final={}
for i in df.columns:
    final[i]={}
    for j in df.columns:
        result = False

        if (i in continious) and (j in continious):
            result = continious_vs_continious(i,j)
        elif (i in continious) and (j in cat):
            result = continious_vs_category(i,j)
        elif (i in cat) and (j in continious):
            result = continious_vs_category(j,i)
        elif (i in cat) and (j in cat):
            result = cat_vs_cat(i,j)

        if result:
            final[i][j]=1
        else:
            final[i][j]=0

final_df = pd.DataFrame(final)
final_df

import seaborn as sns
sns.heatmap(final_df,annot=True,cmap="coolwarm")
```
